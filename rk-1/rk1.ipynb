```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Admission_Predict_Ver1.1.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
sns.violinplot(data=data, y='GRE Score')
```

```python
my_cmap=sns.color_palette('vlag', as_cmap=True)
fig, ax = plt.subplots()
sns.heatmap(data.corr(), cmap=my_cmap, ax=ax, annot=True, fmt='.2f')
```

Целевой признак показывает высокую корреляцию с 'CGPA', 'GRE Score' и 'TOEFL Score', также присутствует корреляция с остальными признаками. Следовательно, модель построить возможно.
При этом признаки, обладающие высокой корреляцией с целевым, также сильно связаны между собой, поэтому, возможно, следует исключить 1 или 2 из них.
